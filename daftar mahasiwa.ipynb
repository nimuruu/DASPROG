```python
def main():
    dataMahasiswa = []
    while True:
        print("\n=== MENU DATA MAHASISWA ===")
        print("1. Tambah Data Mahasiswa")
        print("2. Tampilkan Data Mahasiswa")
        print("3. Keluar")
        try:
            pilihan = input("Pilih menu (1/2/3): ")
            
            if pilihan == "1":
                tambahData(dataMahasiswa)
            elif pilihan == "2":
                tampilkanData(dataMahasiswa)
            elif pilihan == "3":
                print("Terima kasih!!")
                break
            else:
                print("Error: Pilihan harus antara 1, 2, atau 3.")        
        except Exception as e:
            print(f"Terjadi kesalahan: {str(e)}")
            print("Silakan coba lagi.")

def tambahData(dataMahasiswa):
    print("\n=== TAMBAH DATA MAHASISWA ===")
    try:
        while True:
            try:
                nim = input("Masukkan NIM (11 digit): ")
                if not nim.isdigit():
                    raise ValueError("Error: NIM harus berupa angka")
                if len(nim) != 11:
                    raise ValueError("Error: NIM harus 11 digit")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh NIM yang benar: '12345678901'")
        
        while True:
            try:
                nama = input("Masukkan Nama (minimal 6 karakter): ").strip()
                if len(nama) < 6:
                    raise ValueError("Error: Nama harus minimal 6 karakter")
                if not nama.replace(" ", "").isalpha():
                    raise ValueError("Error: Nama harus berupa huruf")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh nama yang benar: 'M Gidzane'")
        
        while True:
            try:
                alamat = input("Masukkan Alamat (minimal 10 karakter): ").strip()
                if len(alamat) < 10:
                    raise ValueError("Error: Alamat minimal 10 karakter")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh alamat yang benar: 'Jl. Cisaat'")
        
        while True:
            try:
                kelas = input("Masukkan Kelas (TI23/TI24): ").upper().strip()
                if kelas not in ["TI23", "TI24"]:
                    raise ValueError("Error: Kelas hanya boleh TI23 atau TI24")
                break
            except ValueError as ve:
                print(ve)
                print("Pilihan kelas: 'TI23' atau 'TI24'")
        
        dataMahasiswa.append({
            "NIM": nim,
            "NAMA": nama,
            "ALAMAT": alamat,
            "KELAS": kelas
        })
        print("Data mahasiswa berhasil ditambahkan!")
        
    except Exception as e:
        print(f"Terjadi kesalahan saat menambah data: {str(e)}")
        print("Silakan coba lagi.")

def tampilkanData(dataMahasiswa):
    print("\n=== DATA MAHASISWA ===")
    try:
        if not dataMahasiswa:
            print("Belum ada data mahasiswa.")
            return
            
        print(f"{'No.':<5}{'NIM':<15}{'NAMA':<20}{'ALAMAT':<25}{'KELAS':<10}")
        print("-" * 75)
        for idx, mahasiswa in enumerate(dataMahasiswa, 1):
            print(f"{idx:<5}{mahasiswa['NIM']:<15}{mahasiswa['NAMA']:<20}{mahasiswa['ALAMAT']:<25}{mahasiswa['KELAS']:<10}") 
            
    except Exception as e:
        print(f"Terjadi kesalahan saat menampilkan data: {str(e)}")

if __name__ == "__main__":
    main()
```
