```python
def main():
    daftarMhs = []
    while True:
        print("\n=== SISTEM INFORMASI MAHASISWA ===")
        print("Silakan pilih menu berikut:")
        print("1. Entri Data Mahasiswa Baru")
        print("2. Lihat Data Mahasiswa")
        print("3. Keluar dari Program")

        try:
            opsi = input("Pilihan Anda 1 / 2 / 3: ")

            if opsi == "1":
                tambahMahasiswa(daftarMhs)
            elif opsi == "2":
                tampilMahasiswa(daftarMhs)
            elif opsi == "3":
                print("Program dihentikan. Terima kasih.")
                break
            else:
                print("Pilihan tidak valid. Harap pilih angka 1, 2, atau 3.")
        except Exception as err:
            print(f"Terjadi kesalahan: {str(err)}")
            print("Silakan coba lagi.")

def tambahMahasiswa(daftarMhs):
    print("\n=== FORMULIR PENDAFTARAN MAHASISWA ===")
    try:
        while True:
            try:
                kodeNIM = input("Tuliskan NIM yang terdiri dari 11 angka: ")
                if not kodeNIM.isdigit():
                    raise ValueError("NIM hanya boleh berisi angka.")
                if len(kodeNIM) != 11:
                    raise ValueError("NIM harus tepat 11 digit.")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh NIM yang benar: 20240040108")

        while True:
            try:
                namaLengkap = input("Tulis nama lengkap mahasiswa: ").strip()
                if len(namaLengkap) < 6:
                    raise ValueError("Nama minimal terdiri dari 6 karakter.")
                if not namaLengkap.replace(" ", "").isalpha():
                    raise ValueError("Nama hanya boleh berisi huruf.")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh nama: Naufal Meitsamar")

        while True:
            try:
                domisili = input("Tulis alamat tempat tinggal: ").strip()
                if len(domisili) < 10:
                    raise ValueError("Alamat minimal 10 karakter.")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh alamat: Nanggeleng, Sukabumi")

        while True:
            try:
                kelasMhs = input("Masukkan kelas (TI23 atau TI24): ").upper().strip()
                if kelasMhs not in ["TI23", "TI24"]:
                    raise ValueError("Kelas hanya boleh TI23 atau TI24.")
                break
            except ValueError as ve:
                print(ve)
                print("Contoh kelas: TI23")

        daftarMhs.append({
            "NIM": kodeNIM,
            "NAMA": namaLengkap,
            "ALAMAT": domisili,
            "KELAS": kelasMhs
        })
        print("Data mahasiswa telah berhasil ditambahkan.")

    except Exception as err:
        print(f"Kesalahan saat menyimpan data: {str(err)}")

def tampilMahasiswa(daftarMhs):
    print("\n=== DAFTAR MAHASISWA TERDAFTAR ===")
    try:
        if not daftarMhs:
            print("Belum ada data yang tersedia.")
            return

        print(f"{'No.':<5}{'NIM':<15}{'NAMA':<20}{'ALAMAT':<25}{'KELAS':<10}")
        print("-" * 75)
        for no, mhs in enumerate(daftarMhs, start=1):
            print(f"{no:<5}{mhs['NIM']:<15}{mhs['NAMA']:<20}{mhs['ALAMAT']:<25}{mhs['KELAS']:<10}")
    except Exception as err:
        print(f"Terjadi kesalahan saat menampilkan data: {str(err)}")

if __name__ == "__main__":
    main()
```
