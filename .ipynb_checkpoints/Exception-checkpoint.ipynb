```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(33)

bulan = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'des']
minggu1 = np.random.randint(200000, 500001, 12)
minggu2 = np.random.randint(200000, 500001, 12)
minggu3 = np.random.randint(200000, 500001, 12)
minggu4 = np.random.randint(200000, 500001, 12)

data = {
    'Bulan' : bulan,
    'Minggu_1' : minggu1,
    'Minggu_2' : minggu2,
    'Minggu_3' : minggu3,
    'Minggu_4' : minggu4
}

df = pd.DataFrame(data)

df['total pengeluaran'] = df[['Minggu_1', 'Minggu_2', 'Minggu_3', 'Minggu_4']].sum(axis=1)

df.head()
```

```python
#Diagram Bar -> proses pembandingan

#Buat Canvas
plt.figure(figsize=(12,6)) #width,height -> inch
plt.bar(df['Bulan'], df['total pengeluaran'], color='skyblue')
plt.title("Pengeluaran Mahasiswa Per Bulan")
plt.xticks(rotation=-45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

for minggu in ['Minggu_1', 'Minggu_2', 'Minggu_3', 'Minggu_4']:
    plt.plot(df['Bulan'], df[minggu], marker='o', label=minggu)

plt.title('Trend Pengeluaran Mahasiswa per Minggu')
plt.xlabel("Bulan")
plt.ylabel("Tren Pengeluaran mahasiswa")
plt.legend()
plt.grid(True)
plt.show
```

```python
#Pie Diagram
plt.figure(figsize=(8,8))
plt.pie(df['total pengeluaran'], labels = df['Bulan'], autopct='%1.1f%%')
plt.show()
```

```python
#Regresi Linear
#Open Source -> scikit -> sklearn

from sklearn.linear_model import LinearRegression

df['Bulan_Angka'] = np.arange(1,13)

x = df[['Bulan_Angka']]
y = df['total pengeluaran']

model = LinearRegression()
model.fit(x,y)

bulan_13 = model.predict([[13]])[0]

print(f"Prediksi Pengeluaran Mahasiswa 2026: {bulan_13:,.0f}")
```
