```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

drivers = pd.read_csv('drivers.csv')

#jumlah pembalap per negera
driver_per_negara = drivers['nationality'].value_counts()

#statitiska dasar
statistik = driver_per_negara.describe()

print("====== STATITISKA DASAR JUMLAH PEMBALAP PER NEGARA=====")
print(f"Jumlah Negara: {statistik['count']} negara")
print(f"Rata-rata (Mean): {statistik['mean']:.2f} pembalap")
print(f"Median: {driver_per_negara.median()} pembalap")
print(f"Standar Deviasi (std): {statistik['std']:.2f} pembalap")
print(f"Minimum: {statistik['min']:} pembalap")
print(f"Maksimum: {statistik['max']:} pembalap")
print(f"Kuartil (1): {driver_per_negara.quantile(0.25)} pembalap")
print(f"Kuartil (3): {driver_per_negara.quantile(0.75)} pembalap")
```

```python
drivers = pd.read_csv('drivers.csv')

# jumlah pembalap per negara dan urutkan
driver_per_country = drivers['nationality'].value_counts().reset_index()
driver_per_country.columns = ['Nationality', 'Jumlah']

# Plot grafik batang
plt.figure(figsize=(16,8))
plt.bar(driver_per_country['Nationality'], driver_per_country['Jumlah'], color='skyblue')

# judul sama label
plt.title('Jumlah Pembalap F1 Berdasarkan Negara (Semua Negara)', fontsize=14)
plt.xlabel('Negara', fontsize=12)
plt.ylabel('Jumlah Pembalap', fontsize=12)

# Rotasi nama negara biar ga bertumpuk
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
